# 파이썬 시작하기!!
* 프로그래밍(programming) : 프로그램을 만드는 것<br>
Pro(미리) + Gram(작성된 것) : 미리 작성된 진행 계획이라 해석..<br><br>

컴퓨터가 이해할 수 있는 값은 오로지 '0'과 '1'로 이루어진 **이진 숫자**이다.<br>
과거에는 이진 숫자를 입력해서 프로그램을 만들었다. 이진 숫자로 이루어진 코드를 **이진 코드**라고 부른다. 하지만 이러한 이진 숫자로 프로그램을 만드는 것은 우리 사람에게는 너무 어렵고 비효율적이라 사람이 이해하기 쉬운 언어로 프로그램을 만드는 것을 목적으로 **프로그래밍 언어**가 만들어졌다.<br>
프로그래밍 언어로 작성한 프로그램을 **소스 코드**라고 한다.<br><br>


---


프로그래밍 언어(사람이 작성) --> 이진 숫자로 변환해주는 코드 실행기 --> 컴퓨터가 인식하고 작동

# 파이썬의 장점
* 문법이 간단해서 배우기 쉽다.<br>
(C언어 등의 다른 프로그래밍 언어보다 문법이 간단하다)
* 많은 사람이 사용하고 있어서 다양한 분야에서 활용할 수 있다.<br>
(기본적인 데스크톱 애플리케이션부터 웹서버, 해킹 도구, IoT, 인공지능 등)
* 대부분의 운영체제(윈도우, 맥, 리눅스)에서 같은 방법으로 사용할 수 있다.

# 파이썬의 단점
* 느리다.<br>
(C언어보다 10~350배 정도 느리다. 프로그래밍 언어는 대부분 '쉽게 사용할 수 있다'라는 성질과 '빠르다'라는 성질이 반비례한다.)<br>
(하지만 최근 컴퓨터 성능이 너무 좋아져서 게임 등과 같이 연산이 많이 필요한 프로그램이 아니라면 프로그래밍 언어의 속도 차이를 크게 느낄 수 없다. 따라서 쉽게 사용할 수 있는 프로그래밍 언어인 파이썬, 루비, 자바스크립트 등 인기를 끌고 있다.)<br><br>

머신러닝, 딥러닝처럼 고정적인 연산을 많이 하는 프로그램은 전체적인 프로그램을 파이썬으로 만들고, 고정적인 연산 부분만 C언어로 만들어 활용하기도 한다.<br><br>

# 파이썬을 배우기 위해 준비해야할 것들
* 개발 환경 : 프로그래밍을 할 수 있는 환경<br>
프로그래밍 언어로 이루어진 코드를 작성할 수 있는 **텍스트 에디터**와 그 코드를 실행할 수 있는 **코드 실행기(파이썬 인터프리터)**가 필요하다.<br><br>

cf)
* **컴파일 언어** : 원시코드(프로그래머가 작성한 소스코드)를 모두 기계어로 변환한 후에 기계(JVM 같은 가상머신)에 넣고 기계어 코드를 실행하는 언어
* **인터프리터 언어** : 원시코드를 기계어로 변환하는 과정없이 한줄 한줄 해석하여 바로 명령어를 실행하는 언어

(반면 **비주얼 스튜디오 코드**는 파이썬 코드 실행기를 포함하지 않기 때문에 통합 개발 환경이 아니고, 단순히 텍스트를 파이썬 문법에 맞게 작성할 수 있도록 도와주는 텍스트 에디터이다.)

# 파이썬 용어들
* 문장 : 실행할 수 있는 코드의 최소 단위
* 프로그램 : 문장들이 모인 것
* 표현식 : 어떠한 값을 만들어 내는 간단한 코드
```
273
10 + 20 + 30 * 10
"Python Programming"
```
* 키워드(대소문자를 구분함)<br>
```
False       None       True     and          as       assert
async       await      break    class     continue     def
del         elif       else     except     finally     for
from        global      if      import       in        is
lambda     nonlocal    not       or         pass      raise
return       try       while    with       yield
```
* 식별자<br>
프로그래밍 언어에서 이름을 붙일 때 사용하는 단어이다.<br>
주로 변수 or 함수 이름 등으로 사용된다.<br>
```
# 식별자 규칙
1. 키워드를 사용하면 안된다.
2. 특수 문자는 언더 바(_)만 허용된다.
3. 숫자로 시작하면 안된다.
4. 공백을 포함할 수 없다.
```
<br>
* 스네이크 케이스 / 캐멀 케이스<br>
1. 스테이크 케이스(snake_case)<br>
단어 사이에 언더 바(_) 기호를 붙여 식별자를 만든다.<br>
  ex) itemlist --> item_list<br>
  함수, 변수 등에 사용<br>

2. 캐멀 케이스(CamelCase)<br>
 단어들의 첫 글자를 대문자로 만들어 식별자를 만든다.<br>
  ex) itemlist --> ItemList<br>
  클래스에서 사용(관례)
<br><br>

* 주석<br>
 프로그램의 진행에 전혀 영향을 주지 않는 코드로, 프로그램을 설명하기 위해 사용한다. (ctrl + /)<br><br>

* 연산자와 자료<br>
 **연산자**는 스스로 값이 되는 것은 아니고 값과 값 사이에 무언가 기능을 적용할 때 사용하는 것을 말한다. 즉 +, -와 같이 단독으로 쓰일 때는 아무 의미를 갖지 못하지만, 숫자와 함께 더하기 빼기 기능을 수행할 때 연산자 기능을 한다.<br>
 **자료**를 리터럴(literal)이라고도 하는데, 숫자이든지 문자이든지 어떠한 값 자체를 의미한다.<br><br>
